```python
import matplotlib.pyplot as plt
import numpy as np

from sklearn.linear_model import LinearRegression
x = np.array([5,15,25,45,35,45,55]).reshape(-1, 1)
y = np.array([5,20,14,32,22,38,10])
z=[]
for i in range(len(x)):
    z.append(x[i]-y[i])
model = LinearRegression()
model.fit(x, y)
y_pred = model.predict(x)
plt.scatter(x, y, color="red", label="actual data")
plt.plot(x, y_pred, color="green", label="Regression line")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Regression ")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

from sklearn.linear_model import LinearRegression
x = np.array([5,15,25,45,35,45,55]).reshape(-1, 1)
y = np.array([5,20,14,32,22,38,10])
z=[]

model = LinearRegression()
model.fit(x, y)
y_pred = model.predict(x)
for i in range(len(x)):
    z.append(abs(y_pred[i]-x[i]))
plt.scatter(x, y, color="red", label="actual data")
plt.plot(x, z, color="green", label="Distance line")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Regression ")
plt.show()
```
